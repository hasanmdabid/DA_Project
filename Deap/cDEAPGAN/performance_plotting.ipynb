```python
""""
This Functon will plot the Epoch vs Losses (Discriminator and generator)
1st we will take the average of d_loss_real , d_loss_fake, and g_loss
2nd we will plot them in the graphs
"""
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
data = pd.read_csv('/home/abidhasan/Documents/DEAPGAN/results/loss_record/valance_D_and_G_loss.csv', sep=';')

data = data.drop(['gen_activation',' batch_per_epoch'], axis=1)
df = data.groupby(by=[" epochs"]).mean()
# Set Seaborn style
sns.set(style="whitegrid")

df.plot(figsize=(12, 8))
plt.xlabel('Epochs', fontsize=18)
plt.ylabel('Loss', fontsize=18)
plt.title('Plot of D_loss and G_loss', fontsize=18)

plt.legend(fontsize=16)

# Rotate x-axis labels for better readability with increased font size
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('/home/abidhasan/Documents/DEAPGAN/picture/DEAP_valance_CGAN_performance_500Epochs.eps', format = 'eps')
```

```python
df_log = df.apply(np.log)
sns.set(style="whitegrid")
# Create the bar plot
plt.figure(figsize=(12, 8))  # Adjust the figure size if needed
df_log.plot()
plt.xlabel('Epochs', fontsize=18)
plt.ylabel('Logarithm of Values', fontsize=18)
plt.title('Plot of Logarithm of D_loss and G_loss', fontsize=18)
#plt.savefig('HARCGAN_performance_100Epochs.jpg')
```
