```python
""""
This Functon will plot the Epoch vs Losses (Discriminator and generator)
1st we will take the average of d_loss_real , d_loss_fake, and g_loss
2nd we will plot them in the graphs
"""
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = pd.read_csv(
    '/home/abidhasan/Documents/HAR/cGAN/D_and_G_loss.csv', sep=';')

data = data.drop(['gen_activation', ' batch_per_epoch'], axis=1)
df = data.groupby(by=[" epochs"]).mean()
df.plot()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Plot of D_loss and G_loss')
plt.savefig(
    '/home/abidhasan/Documents/HAR/figures/cHARGAN_performance_200Epoch.eps', format='eps')
```
